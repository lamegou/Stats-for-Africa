#Projet d'Econométrie de series temporelles



Cette étude a pour principal objectif d'analyser la vente des meubles en Afrique du Sud sur la période 1990 à 2003.

#Analyse des ventes de meubles en Afrique du sud
![](meuble.jpg)

```r
library(readxl)
library(ggplot2)
library(dplyr)
library(lubridate)
library(TSA)
library(caschrono)
library(urca)
proj<-read.csv2(file.choose())

```

```r
str(proj)
```

```r
proj<-data.frame(proj)
proj$Date <- seq(as.Date("1990-01-01"), by = "month", along = proj$Ventes)
```

```r
summary(proj)
```

```r
proj.ts=ts(proj$Ventes,start = c(1990,1),frequency = 12)
par(mfrow=c(2,1))
plot(proj.ts,main="Ventes de meubles")
boxplot(proj.ts,horizontal = T,main='Ventes de meubles',xlab="Ventes")
```

#Analyse des graphiques
Le graphique ci-dessus retrace l'évolution des ventes de meubles sur la période de* 1990 à 2003*. Plusieurs remarques intéressantes peuvent etre faite au sujet de ce graphique.
- En premier lieu, nous notons la présence d'un mouvemennt de fond. Relativement à ce mouvement de fond nous pouvons dire que sur la durée de *1990 à 1997* les ventes ont connu une forte croissance. De *1997 à 2001* une nette baisse des ventes est noté. Cependant cette baisse est de courte durée dans la mesure ou les ventes récommencent à croitre de *2002 à 2003*.
- La chronique des ventes présente un mouvement répétitif oscillatoire de 1990 à 2003. Ce mouvement évoque la présence d'une saisonnalité. En effet  on remarque un pic et un creux chaque année et cela durant toute la période d'étude. La saisonnalité est elle souple ou rigide. Pour répondre à cette question déterminante pour notre étude, nous réaliserons par la suite un graphique des profils saisonniers. Pour l'instant nous retenons un soupçon de saisonnalité.
- La visualisation du boxplot nous permet d'identifier la présence de quelques valeurs atypiques. Ces valeurs atypiques correspondent certainement aux pics observables sur l'évolution de la série.



#Etude de la saisonnalité
L'apercu du graphique de la chronique des ventes nous a permi d'emettre un soupçon relatif à la présence de saisonnalité. Une saisonnalité rigide serait caractéristique d'une saisonnalité significative dans nos données.

#Modélisation automatique du processus générateur des données 

```r
model1<-auto.arima(proj.ts)
summary(model1)
```

```r
summary(model1)
```

```r
res1<-model1$residuals
```

```r
plot(res1,type="o")
```

```r
par(mfrow=c(2,1))
acf(res1,lag.max = 72,lwd=3)

pacf(res1,lag.max =36 ,lwd=3)
```

#Test de Portemanteau
Ce test a pour but d'évaluer la bancheur des résidus.

```r
Box.test.2(x =res1,nlag = 12 )
Box.test.2(x=res1,nlag = 24)
```

_Au vu des p-valeurs , nous pouvons affirmer qu'au seuil de confiance de 5% les résidus issus de la modélisation sont des bruits blancs_

```r
test_stat<-ur.df(y = res1,type = "trend")
summary(test_stat)
```

```r
lag.plot(res1,lag=12)
```

```r
test=ur.df(y = res1,type = "trend",)
```

```r
summary(test)
```
